```python
def options_chain(symbol):
    msft = yf.Ticker(symbol)
    exps = msft.options
    ops = pd.DataFrame()
    for e in exps:
        opt = msft.option_chain(e)
        opt = pd.DataFrame().append(opt.calls).append(opt.puts)
        opt['expirationDate'] = e
        ops = ops.append(opt, ignore_index=True)
    ops['expirationDate'] = pd.to_datetime(ops['expirationDate']) + datetime.timedelta(days = 1)
    ops['dte'] = (ops['expirationDate'] - datetime.datetime.today()).dt.days
    ops['CALL'] = ops['contractSymbol'].str[0:].apply(
        lambda x: "C" in x)
    ops[['bid', 'ask', 'strike']] = ops[['bid', 'ask', 'strike']].apply(pd.to_numeric)
    ops['mark'] = (ops['bid'] + ops['ask']) / 2*1000
    return ops

df = options_chain(str(ticker.upper())).sort_values('dte',ascending=True)

calls = df[df['CALL']==True]
puts = df[df['CALL']==False]
puts = puts.sort_values('expirationDate',ascending = True)
calls = calls.sort_values('expirationDate',ascending = True)


print("enter your desired option's # of days to expiration")
shift = int(input())
print('input the interest rate you would like to use for your black scholes model')
r = float(input()) #2.353
print('input the dividend yield for your selected ticker')
div = float(input()) #0
print("enter your stock's current price")
current_price = float(input())

Tp = float(puts['dte'].reset_index().drop(columns = 'index')['dte'][x])/365
Tc = float(calls['dte'].reset_index().drop(columns = 'index')['dte'][x])/365

Kp = float(puts['strike'].reset_index().drop(columns = 'index')['strike'][x])
Kc = float(calls['strike'].reset_index().drop(columns = 'index')['strike'][x])

S = current_price
S1 =  float((df_['open'] + df_compare_rf['actual']/100*df_['open']).drop(columns = 'index')[days_back])
sigmap = float(puts['impliedVolatility'].reset_index().drop(columns = 'index')['impliedVolatility'][x])
sigmac = float(calls['impliedVolatility'].reset_index().drop(columns = 'index')['impliedVolatility'][x])

#C = price*N( [ln(price/strike)+interest_rate-divYield+vol^2/2)*T]/(vol*sqrt(T)) ] - strike*e^(interest rate * time)*N(d1 - vol*sqrt(T))


def d1(S,K,T,r,div,sigma):
    return(log(S/K)+(r-div+sigma**2/2.)*T)/(sigma*sqrt(T))
def d2(S,K,T,r,div,sigma):
    return d1(S,K,T,r,div,sigma)-sigma*sqrt(T)
def bs_call(S,K,T,r,div,sigma):
    return S*norm.cdf(d1(S,K,T,r,div,sigma))-K*exp(-r*T)*norm.cdf(d2(S,K,T,r,div,sigma))
def bs_put(S,K,T,r,div,sigma):
    return K*exp(-r*T)-S+bs_call(S,K,T,r,div,sigma)
print('Black Scholes Results...')
print(' ---------- ')
print('Price of put: '+str(float((puts['bid'].reset_index().drop(columns = 'index')['bid'][x])+puts['ask'].reset_index().drop(columns = 'index')['ask'][x])/2))
print('Price of call: '+str(float((calls['bid'].reset_index().drop(columns = 'index')['bid'][x])+calls['ask'].reset_index().drop(columns = 'index')['ask'][x])/2))
print(' ---------- ')
print('"immdiate excercise value" of put:'+str(Kp-current_price-float((puts['bid'].reset_index().drop(columns = 'index')['bid'][x])+puts['ask'].reset_index().drop(columns = 'index')['ask'][x])/2))
print('"immdiate excercise value" of call:'+str(current_price-Kc-float((calls['bid'].reset_index().drop(columns = 'index')['bid'][x])+calls['ask'].reset_index().drop(columns = 'index')['ask'][x])/2))
print(' ---------- ')
print('Black-Scholes expected value of put(from stock open today): '+str(bs_put(S,Kp,Tp,r,div,sigmap)))
print('Black-Scholes expected value of call(from stock open today): '+str(bs_call(S,Kc,Tc,r,div,sigmac)))
print('Value of put according to avg of ensemble predictsions in '+str(Tp)[:4]+' day(s): '+str(bs_put(S1,Kp,Tp,r,div,sigmap)))
print('Value of call according to avg of ensemble predictsions in '+str(Tc)[:4]+' day(s): '+str(bs_call(S1,Kc,Tc,r,div,sigmac)))
```
